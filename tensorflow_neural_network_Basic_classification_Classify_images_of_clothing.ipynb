```python
  import tensorflow as tf
  from tensorflow import keras # a high api 
  import numpy as np
  import matplotlib.pyplot as plt
```

```python
data = keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = data.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
train_images = train_images
print(train_images[0])
plt.imshow(train_images[0])
plt.show()
```

```python
model = keras.Sequential([
     keras.layers.Flatten(input_shape=(28 ,28)),                     
     keras.layers.Dense(128, activation="relu"),
     keras.layers.Dense(10, activation="softmax") 
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"]) # metrics: 指標

model.fit(train_images, train_labels, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)

print("Tested Acc:", test_acc)
```

```python
prediction = model.predict(test_images) # this is a 1 D array after converting from this model
print(class_names[np.argmax(prediction[0])]) # argmax return the max value in the array.
plt.imshow(test_images[0])
plt.show()
```

```python
for i in range(5):
  plt.grid(False) # to avoid printing the grid lines
  plt.imshow(test_images[i], cmap=plt.cm.binary)
  plt.xlabel("Actual: " + class_names[test_labels[i]])
  plt.title("Prediction: " + class_names[np.argmax(prediction[i])])
  plt.show()
```
