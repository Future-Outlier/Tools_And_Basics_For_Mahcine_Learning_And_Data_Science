```python
# 0, 1, 2 respectively represent a food.
import numpy as np 
sample = np.array([[[0],[1]],
                  [[1],[1]],
                  [[1],[2]]])
```

```python
print(sample.shape)
```

```python
label = np.array([1, 2, 0])
```

```python
from tensorflow.keras.utils import to_categorical # for one hot encoding
```

```python
sample = to_categorical(sample)
```

```python
print(sample)
```

```python
label = to_categorical(label)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
model = Sequential()
#                         labels          (time, features)
model.add(layers.SimpleRNN(3, input_shape=(2, 3), 
                          activation='softmax' ))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', 
             loss='categorical_crossentropy',
             metrics=['acc'])
```

```python
history = model.fit(sample, label, epochs=500)
```

```python
predict = model.predict(sample)
np.argmax(predict[0]), np.argmax(predict[1]), np.argmax(predict[2])
```
