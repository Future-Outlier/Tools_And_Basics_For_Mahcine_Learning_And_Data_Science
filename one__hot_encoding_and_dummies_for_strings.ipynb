```python
import pandas as pd 
df = pd.read_csv("carprices.csv")
df.head()
```

```python
df.columns = ["Car_Model", "Mileage", "Sell_Prices", "Age"]
```

```python

dummies = pd.get_dummies(df.Car_Model)
dummies.head()
```

```python
merged = pd.concat([dummies, df], axis = 'columns' )
merged.head()
```

```python
final = merged.drop(['Car_Model'], axis = 'columns')
final.head()
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
```

```python
X = final.drop(['Sell_Prices'], axis='columns')
Y = final.Sell_Prices
```

```python
model.fit(X, Y)
predict_parameters = [[1,0,0,50000,1]]
model.predict(predict_parameters)
acc = model.score(X, Y)
print(acc)
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
dfle = df 
dfle.Car_Model = le.fit_transform(dfle.Car_Model)
dfle.head()
```

```python
X = dfle.drop('Sell_Prices', axis="columns")
Y = dfle.Sell_Prices
model.fit(X,Y)
```

```python
acc = model.score(X, Y)
print(acc)
```
