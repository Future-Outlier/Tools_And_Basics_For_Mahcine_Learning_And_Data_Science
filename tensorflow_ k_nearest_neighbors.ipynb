```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
```

```python
data = pd.read_csv("car.data")
print(data.head())
```

```python
# le = preprocessing.LabelEncoder() # le means Label Encoder
# le.fit_transform(values)
# print(le.classes_)
# print(le.transform(values)) # it's encoded by alphabetical order.
# print(le.inverse_transform([0, 1, 2, 3]))
```

```python
le = preprocessing.LabelEncoder() # you can put list into it to encode data from string to integer.
```

```python
# for getting the data to do classification
buying = le.fit_transform(list(data["buying"])) 
print(le.inverse_transform([0,1,2]))
maint = le.fit_transform(list(data["maint"]))
print(le.inverse_transform([0,1,2]))
doors = le.fit_transform(list(data["doors"]))
print(le.inverse_transform([0,1,2]))
persons = le.fit_transform(list(data["persons"]))
print(le.inverse_transform([0,1,2]))
lugboot = le.fit_transform(list(data["lugboot"]))
print(le.inverse_transform([0,1,2]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
print(le.inverse_transform([0,1,2,3]))
```

```python
# split test into testing set and training set
X = list(zip(buying, maint, doors, persons, lugboot, safety))
y = cls
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.1)
```

```python
# to define a classifier
car_classfier = sklearn.neighbors.KNeighborsClassifier(n_neighbors = 5)
car_classfier.fit(X_train, y_train)
acc = car_classfier.score(X_test, y_test) # accruacy
print(acc)
predicted = car_classfier.predict(X_test)
```

```python
values = ['acc' 'good' 'unacc' 'vgood'] # because the alphabetic order
features = ["buying", "maint", "doors", "persons", "lugboot", "safety"]
```

```python
for x in range(len(predicted)): 
  print("Predicted: ", values[predicted[x]], "Label Info: ", X_test[x], "Actual: ", values[y_test[x]])
  n = car_classfier.kneighbors([X_test[x]], n_neighbors=5, True) # Returns neigh_distances and neigh_dist indices
  print("len: ", len(n))
  print("N: ", n)
```
